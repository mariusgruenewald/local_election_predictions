```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
dataframe = pd.read_excel(r'C:\Users\mariu\Documents\Project Local Elections\forecast_candidates.xlsx')
dataframe
```

```python
list_parties = dataframe.party.unique()
list_cities = dataframe.city.unique()
data_sums = []

for j in range(len(list_cities)):
        
    city_now = list_cities[j]
    data_city = dataframe[dataframe['city'].str.match(city_now)]
        
    list_parties = data_city.party.unique()
        
    for k in range(len(list_parties)):
            
        party_now = list_parties[k]
        data_party = data_city[data_city['party'].str.match(party_now)]

        sum_party_city = data_party['Prediction'].sum()
        data_sums.append([sum_party_city, list_parties[k], list_cities[j]])
        
data_sums
```

```python
result_df = pd.DataFrame(data= data_sums, columns = ['sum', 'party', 'city'])

seat_calc_data = pd.DataFrame(columns = [])

    
for j in range(len(list_cities)):
        
    city_now = list_cities[j]
    data_city = result_df[result_df['city'].str.match(city_now)]
        
    list_parties = data_city.party.unique()
        
    for k in range(len(list_parties)):
            
        party_now = list_parties[k]
        data_party = data_city[data_city['party'].str.match(party_now)]
            
        for n in range(1,41):   
                
            number = data_party['sum']/n
                
            num_str = str(number)
                
            number_now = pd.DataFrame([[num_str, n, party_now, city_now]], columns = ['votes_ranked', 'rank',
                                                                                                  'party', 'city'])
                
            seat_calc_data = seat_calc_data.append(number_now, ignore_index = True)

# Something turned awry here. Some in the 'votes_ranked' are 'sum,' and not the number it should be 
# -> if neg estimated vote count
```

```python
calc_2 = seat_calc_data['votes_ranked'].str.split(' ', expand = True)
calc_2 = calc_2[4].str.split('\n', expand = True)
data = pd.concat([calc_2, seat_calc_data], axis = 1)
data.to_excel(r'C:\Users\mariu\Documents\Project Local Elections\seat_evaluation_bavaria.xlsx')
```
