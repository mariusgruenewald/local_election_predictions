```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import quote
from time import sleep
import re
import random


def beautiful_soup(names, year, city):
    
    headers = {
    'User-Agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'
    }
    
    results_next = []
    new_core = []

    for i in range(len(names)):
            
        page_link = 'https://www.ecosia.org/search?q=' + names[i] + ' ' + str(year[i]) + ' '+ city[i]
            
        sleep(random.randint(5,16)) #Script will be stopped for a period of time ranging from 5 to 16 seconds.

        page_response = requests.get(page_link, headers = headers)

        page_content = BeautifulSoup(page_response.content, "html.parser")
        text = page_content.get_text()
        
        core = re.findall(r"\n\n\n\n\n\n\n\n\n\n\n(.*)", text)
        
        new_core.append(core[26])
        
        results_next.append([names[i], year[i], city[i], new_core[i]])
        
    data_google = pd.DataFrame(data=results_next, columns = ['Nachname', 'Jahr', 'Stadt', 'google'])
    
    return data_google
```
