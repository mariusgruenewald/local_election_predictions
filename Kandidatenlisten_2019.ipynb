```python
###############################################################################
############################ Work on dataframe ################################
###############################################################################


def strip_string(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('<div', expand = True)
    return data_new_2

def strip_string_1(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('>', expand = True)
    return data_new_2


def strip_string_2(data, column):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    data_new = data[column].apply(str)
    data_new_2 = data_new.str.split(',', expand = True)
    return data_new_2

def strip_string_3(data, column):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    data_new = data[column].apply(str)
    data_new_2 = data_new.str.split(':', expand = True)
    return data_new_2

def strip_string_4(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('– ', expand = True)
    return data_new_2

def strip_string_5(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('<section', expand = True)
    return data_new_2

def strip_string_6(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('>', expand = True)
    return data_new_2

def strip_string_7(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('<strong>', expand = True)
    return data_new_2

def strip_string_8(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('tbody', expand = True)
    return data_new_2

def strip_string_9(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data['person'].apply(str)
    data_new_2 = data_new.str.split('h2', expand = True)
    return data_new_2


def drop_columns(data, column1, column2):
    data_new = data.loc[:, [column1]]
    data_new['Listenplatz'] = data.loc[:, [column2]]
    data_new.columns = ['Name_total', 'Listenplatz']
    return data_new

def drop_strings_name(data, what, replacement):
    data.Name_total = data.Name_total.str.replace(what, replacement)
    return data
    
def drop_strings_place(data, what, what2, replacement):
    data.Listenplatz = data.Listenplatz.str.replace(what, replacement)
    data.Listenplatz = data.Listenplatz.str.replace(what2, replacement)
    return data

def doctor(data, column):
    
    """ Identify the persons having Dr.
    -> up to now: gives only the name of the person with a doctor title but not only the title"""
    
    new_2 = data[column].str.split('.', expand = True)
    data['Doc'] = new_2[1]
    data.Name_total = data.Name_total.str.replace('Dr.', '').str.replace('med.', ' ').str.replace('Priv.-Doz.', '').str.replace('Prof.', '')
    return data

def general_split(data):
    
    """ split string along '>' 
    -> generalize for any delimiter """
    
    data_new = data.apply(str)
    data_new_2 = data_new.str.split('>', expand = True)
    return data_new_2
```

```python
def google_2(search, time):
    
    results_next = []
    
    for i in range(len(search)):
        name = quote(search[i])
        name = name.replace(' ','+')
        year = quote(time)
        year = year.replace(' ', '+')
        
        page_link = 'https://www.google.com/search?q=' + name + year
            
        sleep(2)

        page_response = requests.get(page_link, timeout=5)

        page_content = BeautifulSoup(page_response.content, "html.parser")

        result = []
    
        for j in range(0,23):
            div = page_content.find_all("div")[j]
            result.append(div)
            
        div_2 = div.find(id="resultStats").text
        results_next.append([search[i], time, div_2])
            
    data_google = pd.DataFrame(data=results_next, columns = ['Name_total', 'Jahr', 'result'])
    
    return data_google
```
