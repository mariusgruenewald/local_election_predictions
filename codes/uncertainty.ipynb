```python
fig = go.Figure()

list_cities = data_graph_5.city.unique()

for i in range(len(list_cities)):
    
    city_now = list_cities[i]
    df_1 = data_graph_5[data_graph_5['city'].str.match(city_now)]
    list_parties = df_1.party.unique()
    
    for j in range(len(list_parties)):
        
        party_now = list_parties[j]
        df_2 = df_1[df_1['party'].str.match(party_now)].reset_index()
        
        council_now = df_2['council_seats'].loc[0]
        color_now = df_2['color'].iloc[0]
        range_color_now = df_2['range_color'].iloc[0]
        _mean = df_2['mean'].iloc[0]
        _min = df_2['lower_limit'].iloc[0]
        _max = df_2['upper_limit'].iloc[0]
        _last = df_2['pre_seats'].iloc[0]
    
        fig.add_trace(go.Indicator(
            domain = {'x': [0, 1], 'y': [0, 1]},
            mode = "gauge+delta",
            value = _mean,
            delta = {'reference': _last,
                     'font': {'family': 'Old Standard TT',
                              'size': 48,
                             },
                     'increasing': {'symbol': '▲'},
                     'decreasing': {'symbol': '▼'},
                    },
            gauge = {'axis': {'range': [None, council_now], # dynamically adjust max to number of seats in coucil
                              'tickmode': 'array',
                              'tickvals': np.round(np.linspace(0, df_2['council_seats'][0],
                                                               num = df_2['council_seats'][0]/3), 
                                                   decimals = 0, out =None),
                              'tickfont': { 'family': 'Old Standard TT'},
                             },
                     'bar': {'color': color_now},
                     'steps' : [
                         {'range': [_min, _max], 'color': range_color_now}
                     ],
                    },
            visible = False,
            title = city_now + ' ' + party_now,

        ))
```

```python
df_2 = data_graph_5.groupby(['city', 'party'], as_index = False).max()
buttons=[]

for i in range(len(df_2.city.unique())):
    
    city_now = df_2.city.unique()[i]
    df_1 = df_2[df_2['city'].str.match(city_now)]
    
    for j in range(len(df_1.party.unique())):
        
        party_now = df_1.party.unique()[j]
        name = city_now + ' ' + party_now
        
        df_2[name] = False
        df_2.loc[(df_2['city'] == city_now) & (df_2['party'] == party_now), name] = True
    
        buttons.append(dict(method='update',
                        label= '<b>' +  name + '</b>',
                        args=[{'visible': df_2[name].values,
                               'title': " <b>" + name + "</b>"}, ],
                           )
                      )
        
buttons
```

```python
fig.layout.update( # Create Buttons dynamically like in other graphs
    updatemenus=[
        go.layout.Updatemenu(
            buttons=buttons,
            y = 1.44,
            x = 0.9,
            bgcolor = '#efefef',
            font = dict(family = 'Old Standard TT',
                        size = 18,
                       ),
            borderwidth = 1.2,
        )
    ]
)
fig['layout'].update(width = 750, height = 375)
filename = 'Uncertaint_6'
py.iplot(fig, filename = filename)
```
