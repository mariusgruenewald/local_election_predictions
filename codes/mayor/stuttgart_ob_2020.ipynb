```python
import pandas as pd
import numpy as np

path = r'C:\Users\mariu\Documents\Project Local Elections/bm_ka_stut.csv'
data_ob = pd.read_csv(path, sep=';')
```

```python
list_supporters = data_ob['support'].astype(str).str.split(',', expand=True)[0].unique()

for i in range(len(list_supporters)):
    
    name = list_supporters[i] + '_support'
    data_ob.loc[data_ob['support'].str.contains(list_supporters[i], regex=False, na=False), name] = 1
    
    data_ob[name] = data_ob[name].fillna(0)

data_ob.head()
```

```python
to_float = ['percent', 'area_industry_pc', 'area_agricul_pc', 'area_forest_pc', 'workers_agricul', 'workers_manu', 'workers_service', 'btw_erststimme_pc']
for z in range(len(to_float)):
    data_ob[to_float[z]] = data_ob[to_float[z]].str.replace(',', '.')
    data_ob[to_float[z]] = data_ob[to_float[z]].astype(float)/100
    
data_ob['age'] = 2020 - data_ob['year_birth'].astype(float)
data_ob = data_ob.rename({'Unnamed:11': 'City_of_birth', 'male':'male_candidate'})
data_ob['visits_total'] = data_ob['visits_total'].str.replace('.','').astype(float)/data_ob['population']
data_ob['visits_foreigner'] = data_ob['visits_foreigner'].str.replace('.', '').astype(float)/data_ob['population']
data_ob['feinstaub'] = data_ob['feinstaub'].str.replace(',', '.').astype(float)/data_ob['population']

to_share = ['male', 'female', 'unter 15', '15-18', '18-25', '25-40', '40-65', '65 u.mehr', 'visits_total', 'visits_foreigner', 'feinstaub', 'workers_agricul', 'workers_manu', 'workers_service',
           'unemployed', 'health_care_receivers', 'asylum_claims']
for u in range(len(to_share)):
    data_ob[to_share[u]] = data_ob[to_share[u]]/data_ob['population']

for column in range(len(data_ob.columns[data_ob.columns.str.contains('kom_wahl')])):
    column_now = data_ob.columns[data_ob.columns.str.contains('kom_wahl')][column]
    data_ob[column_now] = data_ob[column_now].str.replace(',', '.')
    data_ob[column_now] = data_ob[column_now].astype(float)/100

data_ob['size_new_flat'] = data_ob['new_flat_sqm_allowed']*1000/data_ob['numb_new_flat_allowed']
data_ob
```

```python
data_background_name = pd.get_dummies(data_ob['lastname_background_density'])
data_ob = pd.concat([data_ob, data_background_name], axis=1)
data_ob.head()
```

```python
data_ob['job'].unique()
politician = ['MdB', 'MdL', 'OB', 'Bürgermeister', 'Bürgermeisterin', 'Oberbürgermeister', 'Staatsrätin']
business = ['Unternehmerin', 'IT-Unternehmer', 'Medienunternehmer', 'Unternehmer', 'Geschäftsführer', 'Managerin', 'Unternehmensberater']
educ_soc = ['Lehrer', 'Künstler', 'Travestiekünstler\xa0', 'Journalist', 'Gastronom']
laborer = ['Kaufmann', 'Betriebsrat', 'Buchdrucker', 'Drucker', 'EDV-Techniker', 'Industriekaufmann', 'Projektassistentin', 'Logistikmeister']

name_lists = ['politician', 'business', 'educ_soc', 'laborer']
list_of_lists = [politician, business, educ_soc, laborer]

for j in range(len(list_of_lists)):
    list_now = list_of_lists[j]
    for i in range(len(list_now)):
        name = 'job_' + name_lists[j]
        data_ob.loc[data_ob['job'].str.contains(list_now[i], regex=False, na=False), name] = 1
        data_ob[name] = data_ob[name].fillna(0)
        
data_ob.head()
```

```python
# groupby count: candidate, incumbent, runoff (here, number of candidates/incumbents matters)
# groupby mean: 'population', 'pop_density', 'area_industry_pc', 'area_agricul_pc', 'area_forest_pc', 'female', 'unter 15', '15-18', '18-25', '25-40', '40-65', '65 u.mehr', 'male.1', 
# 'workers_agricul', 'workers_manu', 'workers_service', 'unemployed', 'health_care_receivers', 'asylum_claims', 'disp_income', 'investment_p_worker', 'numb_new_flat_allowed', 'new_flat_sqm_allowed', 
# 'visits_total', 'visits_foreigner', 'feinstaub', 
# Strategy group by election (e.g. year and city)

data_ob_runoff_0 = data_ob.groupby(['city', 'year'], as_index=False).count()['candidate']
data_ob_runoff_1 = data_ob.groupby(['city', 'year'], as_index=False).sum()[['city', 'year','runoff', 'incumbent']]
data_ob_runoff_2 = data_ob.groupby(['city', 'year'], as_index=False).mean()[['population', 'pop_density', 'area_industry_pc', 'area_agricul_pc', 'area_forest_pc', 'female', 'unter 15', '15-18', '18-25', 
                                                                             '25-40', '40-65', '65 u.mehr', 'male.1','workers_agricul', 'workers_manu', 'workers_service', 'unemployed', 
                                                                             'health_care_receivers', 'asylum_claims', 'disp_income', 'investment_p_worker', 'numb_new_flat_allowed', 'new_flat_sqm_allowed', 
                                                                             'visits_total', 'visits_foreigner', 'feinstaub']]
data_ob_runoff = pd.concat([data_ob_runoff_1, data_ob_runoff_0, data_ob_runoff_2], axis=1)

data_ob_runoff.loc[data_ob_runoff['runoff'] > 0, 'runoff'] =1
data_ob_runoff
```

```python
data_stuttgart = data_ob[data_ob['city'] == 'Stuttgart']
data_karlsruhe = data_ob[data_ob['city'] == 'Karlsruhe']
```

```python
from sklearn.linear_model import Lasso, SGDClassifier, LogisticRegression # unsure about logreg performance given the case that p > n, omit lasso(lin.reg.)?
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier # likely to perform better since averaging, bagging and other ensemble methods push away from estimation boundaries (0,1) since variance needed to be ~= 0
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB # for poll based forecast
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import StackingClassifier

scaler = StandardScaler()
```

```python
data_ob_runoff_train = data_ob_runoff[data_ob_runoff['year'] != 2020]
data_ob_runoff_test = data_ob_runoff[data_ob_runoff['year'] == 2020]

X_runoff_var = ['population', 'pop_density', 'area_industry_pc', 'area_agricul_pc', 'area_forest_pc', 'female', 'unter 15', '15-18', '18-25', '25-40', '40-65', '65 u.mehr', 'male.1','workers_agricul', 
                'workers_manu', 'workers_service', 'unemployed', 'health_care_receivers', 'asylum_claims', 'disp_income', 'investment_p_worker', 'numb_new_flat_allowed', 'new_flat_sqm_allowed', 
                'visits_total', 'visits_foreigner', 'feinstaub']
y_runoff_var = ['runoff']

X_runoff_train = data_ob_runoff_train[X_runoff_var]
X_runoff_test = data_ob_runoff_test[X_runoff_var]
y_runoff_train = data_ob_runoff_train[y_runoff_var]
```

```python
def combinations_on_off(num_classifiers):
    return [[int(x) for x in list("{0:0b}".format(i).zfill(num_classifiers))]
            for i in range(1, 2 ** num_classifiers)]

param_grid = {'weights':combinations_on_off(7)}
estimator = VotingClassifier(estimators=[('rfc', rfc),('ada', ada),('sgd', sgd),('knnc', knnc),('gbc', gbc),('svc', svc),('logr',log_reg)], voting='hard')
ensemble_cv = GridSearchCV(estimator = estimator, param_grid = param_grid , cv=2, n_jobs = 4, verbose = 10,scoring='f1')
ensemble_cv.fit(X_runoff_train, y_runoff_train)
ensemble_cv.best_params_, ensemble_cv.score(X_runoff_train, y_runoff_train)
```
