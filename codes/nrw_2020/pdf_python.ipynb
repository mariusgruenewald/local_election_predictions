```python
import tabula
import pandas as pd
```

```python
file = r"C:\Users\mariu\Downloads\kw_wahlvorschlaege_bekanntmachung_wahlvorschlaege.pdf"
```

```python
tables = tabula.read_pdf(file, pages = "all", multiple_tables = True,  pandas_options={'header': None})
```

```python
num_tables = range(3, 37) # Have not yet found a way to automatically find until where to loop -> number of election district
master_table = pd.DataFrame()
Wahlkreis_ID = list(range(1,len(num_tables)))

for i in num_tables:
    table = tables[i].dropna()
    table['Walkreis_ID'] = i - 2
    master_table = master_table.append(table)

master_table
```

```python
table_1 = master_table[4].str.split(' ', expand =True)
table_1 = table_1.rename(columns={0: "Plz", 1: "Stadt"})
table_2 = master_table[1].str.split(', ', expand =True)
table_2 = table_2.rename(columns={0: "Nachname", 1: "Vorname"})

j = [table_1, table_2]

for i in [0,1]:
    master_table = pd.concat([master_table, j[i]], axis = 1)
    
master_table = master_table.rename(columns={0: "Kandidaten_ID", 1: "Name_total", 2: "Beruf", 3: "Geburtsjahr", 5: "Partei", 4: "Drop"})
master_table = master_table.drop(["Drop"], axis=1)
master_table
```
