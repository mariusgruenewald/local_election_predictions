```python
import pandas as pd
import numpy as np
dataframe = pd.read_excel(r'C:\Users\mariu\Documents\Project Local Elections\bayern_historisch_migrants.xlsx')
```

```python
dataframe_2 = dataframe[['city', 'year', 'search_result']]
```

```python
list_cities = dataframe_2.city.unique()
list_years = dataframe_2.year.unique()

pre_results_2 = []

for i in range(len(list_cities)):
    
    city_now = list_cities[i]

    for j in range(len(list_years)):
        
        year_now = list_years[j]
        
        dataframe_3 = dataframe_2[(dataframe_2.city == city_now) & (dataframe_2.year == year_now)]
        loc_max = dataframe_3['search_result'].max()
        pre_results = dataframe_3['search_result'].div(loc_max)
        pre_results_2.append(pre_results)

pre_results_3 = pd.DataFrame(data = pre_results_2)
pre_results_4 = pre_results_3.reset_index(drop=True).T

pre_results_4.to_excel(r'C:\Users\mariu\Documents\Project Local Elections\bayern_search_stand.xlsx')
```

```python
dataframe = pd.read_excel(r'C:\Users\mariu\Documents\Project Local Elections\bayern_historisch.xlsx')
```

```python
parties = ['CSU', 'SPD', 'FW', 'Grüne', 'FDP', 'ödp', 'Linke', 'AfD', 'BP', 'REP', 'Piraten']

for k in range(len(parties)):
    
    party = parties[k]
    dataframe[party] = dataframe['party'].str.contains(party, regex=False)
    dataframe.iloc[:,1:] = dataframe.iloc[:,1:].replace({True:1,False:0})
    
dataframe
```

```python
dataframe['youth_list'] = dataframe['party'].str.contains('Ju', regex=False)
dataframe.iloc[:,1:] = dataframe.iloc[:,1:].replace({True:1,False:0})
```

```python
dataframe['local_list'] = 1
dataframe.loc[dataframe['party'] == 'CSU', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'SPD', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'FW', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'Grüne', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'FDP', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'ödp', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'Linke', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'AfD', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'BP', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'REP', 'local_list'] = 0
dataframe.loc[dataframe['party'] == 'Piraten', 'local_list'] = 0
dataframe.loc[dataframe['youth_list'] == 1, 'local_list'] = 0
dataframe
```
