```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
dataframe = pd.read_excel(r'C:\Users\mariu\Documents\Project Local Elections\Bayern_historisch.xlsx')
dataframe_1 = dataframe.dropna().reset_index()
dataframe_1
```

```python
X_subset = dataframe_1[['place_list','search_res_stand', 'doctor', 'Male', 'prof', 'royal', 'muslim', 'incumbent',
                        'non_muslim_migrant', 'CSU', 'SPD', 'Grüne', 'FW', 'Linke', 'FDP', 'ödp', 'BP', 'AfD', 'REP', 'BP', 
                        'Piraten', 'youth_list', 'first_try', 'political_mood']]

Y = dataframe_1[['votes']]

X_scale = dataframe_1[['cost_o_liv']]
```

```python
X_scaled = StandardScaler().fit_transform(X_scale)
scaled_data = pd.DataFrame(data=X_scaled, columns=['Cost_of_liv'])
X = pd.concat((X_subset, scaled_data), axis=1)
```

```python
model = LinearRegression()
reg = model.fit(X, Y)
results = reg.predict(X)
len(results)
```

```python
pred = pd.DataFrame(data=results, columns = ['predicitons'])
results = pd.concat((dataframe_1[['city', 'party', 'year']], pred), axis=1)

list_years = results.year.unique()
data_sums = []

for i in range(len(list_years)):
    
    year_now = list_years[i]
    data_year = results[results['year'] == year_now]
    
    list_cities = data_year.city.unique()    

    for j in range(len(list_cities)):
        
        city_now = list_cities[j]
        data_city = data_year[data_year['city'].str.match(city_now)]
        
        list_parties = data_city.party.unique()
        
        for k in range(len(list_parties)):
            
            party_now = list_parties[k]
            data_party = data_city[data_city['party'].str.match(party_now)]

            sum_party_city = data_party['predicitons'].sum()
            data_sums.append([sum_party_city, list_party[k], list_cities[j], list_years[i]])
```

```python
result_df = pd.DataFrame(data= data_sums, columns = ['sum', 'party', 'city', 'year'])

seat_calc = []

for i in range(len(list_years)):
    
    year_now = list_years[i]
    data_year = result_df[result_df['year'] == year_now]
    
    list_cities = result_df.city.unique()
    
    for j in range(len(list_cities)):
        
        city_now = list_cities[j]
        data_city = data_year[data_year['city'].str.match(city_now)]
        
        list_parties = data_city.party.unique()
        
        for k in range(len(list_parties)):
            
            party_now = list_parties[k]
            data_party = data_city[data_city['party'].str.match(party_now)]
            
            for n in range(1,60):   
                
                number = data_party['sum']/n
                
                number_str = str(number)
                
                seat_calc.append([number_str, n, party_now, city_now, year_now])
```

```python
calc = pd.DataFrame(data= seat_calc, columns = ['calculus', 'seat_theoretical', 'party', 'city', 'year'])
calc_2 = calc['calculus'].str.split(' ', expand = True)
calc_2[4] = calc_2[4].str.split('\n', expand = True)
calc['calculus'] = calc_2[4]
calc.to_excel(r'C:\Users\mariu\Documents\Project Local Elections\seat_evaluation_bavaria.xlsx')
```
