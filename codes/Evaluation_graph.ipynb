```python
for j in range(len(cities)):
    
    city_now = cities[j]
    data_trace = dataframe_1[dataframe_1['city'].str.match(city_now)]
    data = []
    
    for i in range(len(parties)):
        
        party_now = parties[i]
        color_now = color[i]
        trace = go.Scatter(
            x = data_trace[data_trace['party'].str.match(party_now)]['votes_pred'],
            y = data_trace[data_trace['party'].str.match(party_now)]['votes'],
            mode='markers',
            name = party_now,
            marker=dict(
                color=color_now,
                size=12,
                line=dict(
                    color='rgba(217, 217, 217, 0.14)',
                    width=0.5),
                opacity=0.8)
        )
        data.append(trace)

    trace_1 = go.Scatter(
        x = [0, 1000, 2000, 5000, 10000, 15000, 20000, 23000],
        y = [0, 1000, 2000, 5000, 10000, 15000, 20000, 23000],
        name = 'Fit',
        mode = 'lines',
        marker = dict(
            color = 'black')
    )

    data.append(trace_1)
    
    filename = city_now + ' ' + 'Forecast'
    
    layout = go.Layout(
        legend=dict(orientation="h")
    )
    fig = dict(data=data, layout = layout)
    fig['layout'].update(width = 800, height = 800)
    fig['layout'].update(title = filename)
    fig['layout'].update(xaxis = dict(
                    title = 'Predicted Votes'))
    fig['layout'].update(yaxis = dict(
                    title = 'Actual Votes',
                    tickmode = 'auto'))

    py.iplot(fig, filename = filename)
```
